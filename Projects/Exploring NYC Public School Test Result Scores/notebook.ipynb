```python
#Imports
import pandas as pd
```

```python
#Load the Data
schools = pd.read_csv("Dataset/schools.csv")
schools.head()
```

```python
#Best Math Schools in Descending Order
best_math_schools = schools[schools["average_math"] >= 0.8*800][["school_name","average_math"]].sort_values(by="average_math",ascending=False)
best_math_schools.head()
```

```python
#Top 10 schools
schools["total_SAT"] = schools["average_math"] + schools["average_reading"] + schools["average_writing"]
top10_schools = schools[["school_name","total_SAT"]].sort_values(by="total_SAT",ascending=False)
top10_schools.head()
```

```python
#Large Standart Deviation Borough
largest_std_dev= schools.groupby("borough").agg({"total_SAT":["count","mean","std"]})
largest_std_dev = largest_std_dev.rename(columns={"count":"num_schools","mean":"average_SAT","std":"std_SAT"})
```
