```python
#Imports
import pandas as pd
```

```python
#Loading Datasets
employee_information_header = ["employee_id", "last_name", "first_name", "emergency_contact", "emergency_contact_number","relationship"]
employee_information = pd.read_excel("Dataset/employee_information.xlsx", sheet_name="emergency_contacts",header=None,names=employee_information_header) #header = None make the data that are in the header part of the data rows, names pass the names of the columns
employee_addresses = pd.read_excel("Dataset/employee_information.xlsx")
office_addresses = pd.read_csv("Dataset/office_addresses.csv")
roles = pd.read_json("Dataset/employee_roles.json",orient="index").reset_index(names="employee_id")#We use orient = "index" because pandas when read json files need this to make the table right depending on the json format and we reset_index to turn employee_id a colummn that we need to merge
```

```python
#Merging
employees = employee_information.merge(employee_addresses, on = "employee_id", how="outer")
employees = employees.drop(columns=["employee_last_name","employee_first_name"])
employees_temp = employees.merge(office_addresses, left_on="employee_country",right_on="office_country",how="outer")
employee = employees_temp.merge(roles,on="employee_id",how="outer")
```

```python
#Filling NaN values and set the index to be employee_id
employee_final = employee.fillna("Remote").set_index("employee_id")
employee_final.head()
```
